```python
import pandas as pd # 판다스 임포트.
import numpy as np
import matplotlib.pyplot as plt # 맷플롯립 임포트.
import seaborn as sns # 시본 임포트
import csv
import yfinance as yf
```

```python

data_x = pd.read_csv('/content/drive/MyDrive/SWEnglish/^IXIC_daily20152022.csv', sep = ',') #/content/drive/MyDrive/SWEnglish/IXIC_month.csv
data_x['Open'].shape
data_z = list()
for i in range(len(data_x)):
  if(i == len(data_x) - 1):
    data_z[i] = 0
  else:
    data_z[i] = data_x['Open'][i + 1] / data_x['Open'][i]

data_y = pd.read_csv('/content/drive/MyDrive/SWEnglish/FFR20152022.csv', sep = ',') #/content/drive/MyDrive/SWEnglish/FFR_month.csv
data_y['DFF'].shape

X = data_x.Close.values
Y = data_y.DFF.values

plt.scatter(X, Y, alpha=0.5)
plt.title('Nasdaq - FED relation (2015~2022)') #두 지표의 상관 관계 분석 가능
plt.xlabel('Nasdaq')
plt.ylabel('FFR')
plt.show()

cov = (np.sum(X*Y) - len(X)*np.mean(X)*np.mean(Y)) / len(X)
print(cov)
```

```python
corr = cov / (np.std(X) * np.std(Y)) ## corr 값이 0.3~0.7(뚜렷한 양적 선형관계), 0.7~1(강한 양적 선형관계)
print(corr)
```
